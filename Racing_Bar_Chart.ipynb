```python
import pandas as pd
import plotly.express as px
import plotly
```

```python
df = pd.read_excel('/content/drive/MyDrive/Racing Bar Chart/Racing Bar Chart .xlsx')
```

```python
df
```

```python
country = df['Country']
sales = df['Sales']
date = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
fig = px.bar(df, x = country,y = sales,color=country,animation_frame= date,animation_group= country,range_y = [0,1200])
```
